```python
#Extracting first character from ticket 
def extract(ticket):
    Ticket=ticket
    
    if Ticket[0].isalpha() or int(Ticket[0])>=3:
        return 3
    elif int(Ticket[0])==2:
        return 2
    elif int(Ticket[0])==1:
        return 1
```

```python
train['Ft']=train['Ticket'].apply(extract)
```

```python
def fill_age(col):
    Age=col[0]
    Title=col[1]
    
    if pd.isnull(Age):
        if Title==0:
            return 32
        elif Title==1:
            return 21
        elif Title==2:
            return 35
        elif Title==3:
            return 4
        elif Title==4:
            return 44
        elif Title==5:
            return 37
    else:
        return Age
```

```python
train['Age']=train[['Age','Title']].apply(fill_age,axis=1)
```

```python
train.info()
```

```python
train['Embarked'].fillna(value=0,inplace=True)
```

```python
TRAIN=train.drop(['Name','Ticket','Cabin'],axis=1)
```

```python
TRAIN
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X=TRAIN.drop(['PassengerId','Survived'],axis=1)
y=TRAIN['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
predictions=lg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```
